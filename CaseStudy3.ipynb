```python
import numpy as np
import pandas as pd

terror = pd.read_csv("D:/Python/Python Assignments/DataSets/terror.csv", sep=',', header=0, encoding = 'Latin')
```

```python
# Checking null values in Eventid column
terror.eventid.isnull().sum()
```

```python
# Q.1)	How many attacks happened in India?
attack = terror[terror['country_txt'] == 'India'].eventid.count()
print(attack, 'attacks happened in India!')
```

```python
# Q.2)	How many attacks happened in India and upto 3 people were killed? 
kill = terror[(terror['country_txt'] == 'India') & (terror['nkill'] <= 3)].eventid.count()
print(kill, 'attacks happened in India and upto 3 people were killed!')
```

```python
# Q.3) Extract the city and summary for attacks above
terror[["city", "summary"]][(terror['country_txt'] == 'India') & (terror['nkill'] <= 3)]
```

```python
# Q.4)	In a single terror incident in India, find out top 5 cities by number killed
terror[['eventid', 'city', 'nkill']][terror['country_txt'] == 'India'].sort_values('nkill', ascending=False).head(5)
```

```python
# Q.5)	In a single terror incident in India, find out top 5 cities by number killed and wounded 
terror[terror.country_txt=='India'].sort_values(by=['nkill'], ascending=False)[['city','nkill','nwound']].head(5)
```

```python
# Q.6) How many attacks were successful that were suicide attacks?
suc = terror[(terror['suicide'] == 1) & (terror['success'] == 1)].eventid.count()
print('There were',suc,'successful attacks that were suicide attacks.')
```

```python
# Q.7) label all the incidents where the number killed was more than 5 as severe. 
terror['label'] = np.where(terror['nkill'] > 5, 'severe', 'notsevere')

```

```python
# Q.8) write a function to label an incident that was both successful and suicidal
def check(event) :
    if ((terror['suicide'][terror['eventid'] == event] == 1).tolist()[0] & 
        (terror['success'][terror['eventid'] == event] == 1).tolist()[0]) :
        print('Yes, the incident that was both successful and suicidal.')
    else :
        print('No, the incident that was both not successful nor suicidal.')
```

```python
# Q.11)	List the number of suicides attacks and average kills by Af-Pak-India vs ROW. 
# Rename columns in the output as Average_Kills and Number_Incidents.
temp = terror[terror.suicide == 1].groupby('category').agg({'suicide':'count', 'nkill':'mean'})
temp = temp.rename(columns={'suicide': 'Number_Incidents', 'nkill': 'Average_Kills'})
temp
```
